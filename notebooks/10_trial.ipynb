```python
df_1 = pd.read_csv('..\\artifacts\\data\\raw\\train_data1.csv')
df_2 = pd.read_csv('..\\artifacts\\data\\raw\\train_data2.csv')
df_3 = pd.read_csv('..\\artifacts\\data\\raw\\train_data3.csv')
```

```python
test_data_merger_1 = pd.merge(left = df_1,
                              right = df_2,
                              how = 'outer',
                              on = 'ident_id')
test_data_merger_final = pd.merge(left = test_data_merger_1,
                                  right = df_3,
                                  how = 'outer',
                                  on = 'ident_id')
test_data_merger_final_copy = test_data_merger_final.copy()
```

```python
import numpy as np
test_data_merger_final_copy.drop(columns='ident_id',inplace=True)
test_data_merger_final_copy.rename(columns={'field_74_':'class'},inplace=True)
test_data_merger_final_copy['class'] = test_data_merger_final_copy['class'].map({'neg':0,'pos':1})
test_data_merger_final_copy.replace('na',np.nan,inplace=True)
test_col_list = [i for i in test_data_merger_final_copy.columns if i != 'class']
for i in test_col_list:
    test_data_merger_final_copy[i]=test_data_merger_final_copy[i].astype('float')
```

```python
check_df = test_data_merger_final_copy.drop(columns='class').describe()==df.drop(columns='class').describe()
for i in check_df.columns:
    print(i,check_df[i].unique())
```
