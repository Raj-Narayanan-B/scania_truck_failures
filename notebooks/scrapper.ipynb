```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = "https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html"
```

```python
# Get the html page using url
page = requests.get(url)
```

```python
#Display the html page
page.content
```

```python
# reading html file page.content 
soup = bs(page.content, 'html.parser')
soup
```

```python
# Prettify() function in beautiful soup will enable to view the html page with how tags are nested in document / page
print(soup.prettify())
```

```python
names = soup.find_all('span', class_ = 'classifier')
names
```
